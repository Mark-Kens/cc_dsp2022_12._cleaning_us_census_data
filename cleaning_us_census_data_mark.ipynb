2. It will be easier to inspect this data once we have it in a DataFrame. You can't even call `.head()` on these `csv`s! How are you supposed to read them?

   Using `glob`, loop through the census files available and load them into DataFrames. Then, concatenate all of those DataFrames together into one DataFrame, called something like `us_census`.

````python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import glob
# csvs = glob.glob("states*.csv")
csvs = sorted(glob.glob("states*.csv"))

dataframes = []
for i in csvs:
    dataframe_i = pd.read_csv(i)
    dataframes.append(dataframe_i)
# print(dataframes)
# The list of states in "dataframe" is not alphabetically sorted.
# Reason: glob did not open them in order.
# Reason: The operating system sorst the files in the directory with some unkonwn logic.
#         Using "ls -U" (do not sort) returns the files in the directory in seemingly arbitrary order.
#         Files are not sorted by name or size or date.

# Solution: use "sorted" with glob.

us_census = pd.concat(dataframes)
print(us_census.head())
````

## Regex to the Rescue

5. Use regex to turn the `Income` column into a format that is ready for conversion into a numerical type.

````python
# Trying to access values in the "Income" column with the index (us_census['Income'][0]) did not work. 
# Need to use dataframe.iat: Access a single value for a row/column pair by integer position.

for i in range(0, len(us_census['Income'])):
    us_census['Income'].iat[i] = str(us_census['Income'].iat[i]).replace('$', '').replace(',', '')
    # Each value in "Income"   = convert to string; replace "$"; replace ","

us_census["Income"] = pd.to_numeric(us_census['Income'])

us_census['Income']
````

6. Look at the `GenderPop` column. We are going to want to separate this into two columns, the `Men` column, and the `Women` column.

   Split the column into those two new columns using `str.split` and separating out those results.

````python
Men = []
Women = []
for i in range(0,len(us_census["GenderPop"])):
    converted_to_str = str(us_census['GenderPop'].iat[i])
    # Convert value to string
    split_items_list = converted_to_str.split('_')
    Men.append(split_items_list[0])
    Women.append(split_items_list[1])

us_census['Men'] = Men
us_census['Women'] = Women

us_census.head()
````

7. Convert both of the columns into numerical datatypes.

   There is still an `M` or an `F` character in each entry! We should remove those before we convert.

````python
for i in range(0,len(us_census["Men"])):
    converted_to_str = str(us_census['Men'].iat[i])
    edit_on_string = converted_to_str.replace('M', '')
    us_census['Men'].iat[i] = edit_on_string
    
for i in range(0,len(us_census["Women"])):
    converted_to_str = str(us_census['Women'].iat[i])
    edit_on_string = converted_to_str.replace('F', '')
    us_census['Women'].iat[i] = edit_on_string
    
us_census['Men'] = pd.to_numeric(us_census['Men'])
us_census['Women'] = pd.to_numeric(us_census['Women'])

us_census.head()
````

````python
# Requirements: 1. The first visualization your boss wants you to make is a scatterplot 
# that shows average income in a state vs proportion of women in that state.

# Average income in a state = "Income" column

# New column for proportion of women in a state: prop_women = Women / TotalPop
us_census['prop_women'] = ""
us_census['prop_women'] = us_census.Women / us_census.TotalPop

plt.scatter(us_census['prop_women'], us_census['Income'])
plt.title("Scatter Plot of Income vs. Proportion of Women per State")
plt.xlabel("Proportion of Women per State")
plt.ylabel("Income (in US Dollars)")
plt.show()
plt.clf()

us_census.head(30)
# There are NaN values in the data. 

````

9. You want to double check your work. You know from experience that these monstrous csv files probably have `nan` values in them! Print out your column with the number of women per state to see.

   We can fill in those `nan`s by using pandas' `.fillna()` function.
   
   You have the `TotalPop` per state, and you have the `Men` per state. As an estimate for the `nan` values in the `Women` column, you could use the `TotalPop` of that state minus the `Men` for that state.
   
   Print out the `Women` column after filling the `nan` values to see if it worked!

````python
us_census['Women'] = us_census['Women'].fillna(us_census['TotalPop'] - us_census['Men'])
print(us_census['Women'])

us_census['prop_women'] = ""
us_census['prop_women'] = us_census.Women / us_census.TotalPop

plt.scatter(us_census['prop_women'], us_census['Income'])
plt.title("Scatter Plot of Income vs. Proportion of Women per State")
plt.xlabel("Proportion of Women per State")
plt.ylabel("Income (in US Dollars)")
plt.show()
plt.clf()

# New data points appeared in the plot.
````
